### B. 명령어 처리 (예: /start, /help 명령어)

1. get_updates(offset=None): 봇이 계속해서 메시지를 가져오고, offset을 사용하여 중복 메시지를 받지 않도록 설정
2. handle_commands(chat_id, message_text): 명령어를 처리하는 함수로, 사용자가 /start 또는 /help 같은 명령어를 입력하면 이에 맞게 응답하고 나머지 일반 메시지는 그대로 "You said: {message_text}" 형태로 응답하도록 설정
3. run_bot(): 봇이 계속해서 메시지를 수신하고 처리하는 메인 루프로서 일정 시간마다 업데이트를 확인하고 새로운 메시지를 처리합니다.

```python
import urllib3
import json
import time

BOT_TOKEN = "8173266600:AAFsrCXFDYSwrOLvdhyR0jbYDil1F3seq0w"

# Telegram 봇 API로부터 업데이트를 가져오는 함수
def get_updates(offset=None):
    http = urllib3.PoolManager()
    url = f"https://api.telegram.org/bot{BOT_TOKEN}/getUpdates?timeout=100"
    if offset:
        url += f"&offset={offset}"
    response = http.request('GET', url)
    return json.loads(response.data.decode('utf-8'))

# 사용자에게 메시지를 보내는 함수
def send_message(chat_id, text):
    http = urllib3.PoolManager()
    url = f"https://api.telegram.org/bot{BOT_TOKEN}/sendMessage"
    message = {
        'chat_id': chat_id,
        'text': text
    }
    encoded_message = json.dumps(message).encode('utf-8')
    response = http.request('POST', url, body=encoded_message, headers={'Content-Type': 'application/json'})
    return response.data.decode('utf-8')

# 명령어 및 키워드 처리
def handle_commands(chat_id, message_text):
    if message_text.lower() == "/start":
        send_message(chat_id, "Welcome to the bot! How can I assist you today?")
    elif message_text.lower() == "/help":
        help_text = (
            "/start - Start the bot and get a welcome message\n"
            "/help - Get a list of commands\n"
            "You can also ask me general questions!"
        )
        send_message(chat_id, help_text)
    else:
        send_message(chat_id, f"You said: {message_text}")

# 실시간으로 봇이 대화 수신 및 처리
def run_bot():
    last_update_id = None

    while True:
        updates = get_updates(last_update_id)

        if 'result' in updates and len(updates['result']) > 0:
            for update in updates['result']:
                # 가장 최신 메시지 정보 가져오기
                message = update['message']
                chat_id = message['chat']['id']
                message_text = message.get('text', '')

                # 명령어 처리
                handle_commands(chat_id, message_text)

                # 최신 업데이트 ID 저장
                last_update_id = update['update_id'] + 1
        time.sleep(1)

if __name__ == "__main__":
    run_bot()
```
